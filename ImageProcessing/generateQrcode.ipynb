```python
import numpy as np
import qrcode
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
Generate = qrcode.make("https://www.qr-code-generator.com/")
```

```python
Generate.save("ali.png")
generated_qr_code_cv = np.array(Generate)
```

```python
myimage = cv.imread("ali.png")
myimage = cv.cvtColor(myimage, cv.COLOR_BGR2GRAY)
myimage = cv.cvtColor(myimage, cv.COLOR_GRAY2RGB)
plt.imshow(myimage)
```
